```python
%sql
-- Vista de análisis con categorías
CREATE OR REPLACE VIEW wine_analysis_view AS
SELECT *
FROM wine_quality_clean;

SELECT COUNT(*) AS total FROM wine_analysis_view;
```

```python
%sql
-- Ranking de vinos por alcohol promedio y calidad
SELECT
  wine_type,
  quality,
  ROUND(AVG(alcohol), 2)          AS avg_alcohol,
  ROUND(AVG(volatile_acidity), 3) AS avg_volatility,
  ROUND(AVG(sulphates), 3)        AS avg_sulphates,
  COUNT(*)                         AS total_wines
FROM wine_quality_clean
GROUP BY wine_type, quality
ORDER BY wine_type, quality DESC;
```

```python
%sql
-- ¿Cuáles características tienen los vinos de calidad alta?
SELECT
  quality_category,
  wine_type,
  ROUND(AVG(alcohol), 2)          AS avg_alcohol,
  ROUND(AVG(volatile_acidity), 3) AS avg_volatile_acid,
  ROUND(AVG(citric_acid), 3)      AS avg_citric_acid,
  ROUND(AVG(sulphates), 3)        AS avg_sulphates,
  ROUND(AVG(residual_sugar), 3)   AS avg_sugar,
  COUNT(*)                         AS cantidad
FROM wine_analysis_view
GROUP BY quality_category, wine_type
ORDER BY quality_category DESC, wine_type;
```

```python
%sql
-- Detectar vinos "atípicos": alcohol muy alto o muy bajo
SELECT *
FROM wine_quality_clean
WHERE alcohol > (SELECT AVG(alcohol) + 2 * STDDEV(alcohol) FROM wine_quality_clean)
   OR alcohol < (SELECT AVG(alcohol) - 2 * STDDEV(alcohol) FROM wine_quality_clean)
ORDER BY alcohol DESC;
```

```python
%sql
-- Distribución de densidad por tipo de vino
SELECT
  wine_type,
  ROUND(MIN(density), 4)  AS min_density,
  ROUND(AVG(density), 4)  AS avg_density,
  ROUND(MAX(density), 4)  AS max_density,
  ROUND(STDDEV(density), 5) AS std_density
FROM wine_quality_clean
GROUP BY wine_type;
```
